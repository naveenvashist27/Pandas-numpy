```python
import numpy as np
import pandas as pd

data = {
    "Voter_id": range(1001,1051),
    "Age": np.random.randint(18,80,50),
    'Gender': np.random.choice(['Male', 'Female', 'Other'], 50),
    'State': np.random.choice(['MP', 'UP', 'MH', 'RJ', 'DL'], 50),
    'Voted_For': np.random.choice(['BJP', 'Congress', 'AAP'], 50),
    'Voting_Status': np.random.choice(['Voted', 'Not Voted'], 50)
}
df = pd.DataFrame(data)
df
```

```python
# find data which have age greater than 40
df['Age']>=40                                               #this will result in true false 
df[df['Age']>=40]                                           #this will give data of those having age >=40
df[(df['Age']>=40) & (df['Age']<=70)]                       # for age>=40 & <=70
len(df[(df['Age']>=40) & (df['Age']<=70)])                  # for number of peoples between age>=40 & <=70
df['Age'].between(40,70)                                    #this will result in true false 
sum(df['Age'].between(40,70))                               #this will result in number of peoples
df[df['Age'].between(40,70)]                                #this will result data of peoples
```
